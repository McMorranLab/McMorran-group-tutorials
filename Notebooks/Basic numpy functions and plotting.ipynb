```python
# make sure all the plots show up in the notebook rather than in a separate window
%matplotlib inline 
from matplotlib import pyplot as plt # this is a 2d plotting library and comes with Anaconda
import numpy as np # NumPy is the fundamental package for scientific computing with Python
```

Scientific computing and plotting data go hand in hand here are a few ways to plot the function $f(x) = sin(x)$ from $0$ to $4 \pi$.

```python
# first we need to define x here a way
# we define x as a python list of containing
# 101 numbers equally spaced between 
x = []
for i in range(100+1):
    x.append(i*4*np.pi/100.)
```

```python
for i in x:
    print(i)
```

```python
# now we can define y
y = []
for i in x:
    y.append(np.sin(i))
```

```python
# and here we can plot it
plt.plot(x,y)
plt.show()
```

That worked but it was way harder than it needed to be here are two more ways to do it.

```python
xx = np.linspace(0, 4*np.pi, 101) # this makes a numpy array with 101 numbers equally spaced between
# 0 and 4*pi (start, stop, number of samples)
xxx = np.arange(0, 4*np.pi+4*np.pi/101., 4*np.pi/100.) # this is another way of doing the same thing
# (start, stop, step size)
```

```python
# plot all 3
plt.plot(x,y)
plt.show()

plt.plot(xx,np.sin(xx))
plt.show()

plt.plot(xxx,np.sin(xxx))
plt.show()
```

```python
# or we could plot them all on the same plot
# I am adding 0.25 to each so they can each
# be seen
plt.plot(x,y, label='first')
plt.plot(xx,np.sin(xx)+0.25, 'o', label='second')
plt.plot(xxx,np.sin(xxx)+0.5, 's-', label=r'$3^{rd}$')
plt.legend()
plt.show()
```

We can do the same thing with a bunch of other functions that are built into numpy.

```python
plt.figure(figsize=(15,7))
plt.plot(x,y, label='first')
plt.plot(xx,np.cos(xx), 'o', label='second')
plt.plot(xxx,np.tan(xxx), 's-', label=r'$3^{rd}$')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Some trigonometric functions')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,7))
x = np.linspace(-20,20,100)
y = 30*x + 5
plt.plot(x,y, label='first')
y = y - 5.*x**2
plt.plot(x,y, 'o', label='second')
y = y + (0.7*x)**3
plt.plot(x,y, 's-', label=r'$3^{rd}$')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Some polynomial functions')
plt.legend(loc='best')
plt.show()
```

Numpy also has many other useful functions including several random number generators.

```python
#this is a plot of 'noise' 100 random numbers uniformly distributed in the interval [0,1)
# see http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.rand.html
plt.plot(np.random.rand(100))
```

We can check to make sure this really has a uniform distribution using a histogram.

```python
plt.hist(np.random.rand(100000), 20, normed=1)
plt.ylabel('random number distribution')
plt.show()
```

Or we can us a different random number generator that draws numbers from the normal or "Gaussian distribution".

```python
plt.hist(np.random.randn(100000), 50, normed=1)
plt.ylabel('random number distribution')
plt.grid(True)
plt.show()
```
