```python
x = np.linspace(-5,5,1000)
k = np.fft.fftfreq(len(x), d=(x[1]-x[0]))
plt.plot(x,(np.abs(x)<2))
plt.ylim([-0.1,1.1])
plt.show()
plt.plot(k, (np.abs(k)<0.5)*np.abs((x[1]-x[0])*np.fft.fft(np.abs(x)<2)))
plt.ylim([0,0.5])
plt.xlim([-5,5])
plt.show()
plt.plot(x, np.real(np.fft.ifft((np.abs(k)<0.5)*np.fft.fft(np.abs(x)<2))))
plt.plot(x, np.real(np.fft.ifft((np.abs(k)<1.7)*np.fft.fft(np.abs(x)<2))))
#plt.ylim([-0.1,1.1])
plt.xlim([-5,5])
plt.show()
```
