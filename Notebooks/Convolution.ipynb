```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
square_wave1=np.repeat([0.,0.,0.,0.,1.,0.,0.],1000)
square_wave2=np.repeat([0.,1.,1.,1.,0.],100)
plt.plot(square_wave1)
plt.ylim([-0.1,2])
plt.show()
plt.plot(square_wave2)
plt.ylim([-0.1,2])
plt.show()
```

```python
convolution=signal.convolve(square_wave1, square_wave2, mode='same')
#plt.ylim([0,110])
plt.plot(convolution)
```
